```python
import numpy as np
from scipy.integrate import dblquad

# Verbesserte Funktion zur Berechnung der Schnittfläche zweier Kreise
def overlap_area(r1, r2, d=1):
    if d >= r1 + r2:  # Kreise überschneiden sich nicht
        return 0
    elif d <= abs(r1 - r2):  # Ein Kreis ist vollständig im anderen enthalten
        return np.pi * min(r1, r2)**2
    
    # Berechnung der Schnittfläche, wenn die Kreise sich teilweise überschneiden
    part1 = r1**2 * np.arccos((d**2 + r1**2 - r2**2) / (2 * d * r1))
    part2 = r2**2 * np.arccos((d**2 + r2**2 - r1**2) / (2 * d * r2))
    part3 = 0.5 * np.sqrt((-d + r1 + r2) * (d + r1 - r2) * (d - r1 + r2) * (d + r1 + r2))
    return part1 + part2 - part3

# Funktion f(x, y) zur Berechnung der Fläche der beiden Viertelkreise abzüglich der Überlappung
def f(x, y):
    # Radien r1 und r2 zum Punkt (x, y)
    r1 = np.sqrt(x**2 + y**2)
    r2 = np.sqrt((1 - x)**2 + y**2)
    
    # Fläche der beiden Viertelkreise
    quarter_circle_area1 = np.pi * r1**2 / 4
    quarter_circle_area2 = np.pi * r2**2 / 4
    
    # Berechnung der Überlappungsfläche
    A = overlap_area(r1, r2)
    
    # Berechnung der Gesamtfläche f(x, y)
    return quarter_circle_area1 + quarter_circle_area2 - A 

# Definition des Integranden
def integrand(y, x):
    return f(x, y)

# Integration über die beiden Bereiche des Dreiecks
mean_area1, _ = dblquad(integrand, 0, 0.5, lambda x: 0, lambda x: x)
mean_area2, _ = dblquad(integrand, 0.5, 1, lambda x: 0, lambda x: 1 - x)

# Gesamtes Integral
total_integral = mean_area1 + mean_area2

# Berechnung des Durchschnitts durch Division mit der Fläche des Dreiecks (0.25)
mean_area = total_integral / 0.25
print(f"Exakter durchschnittlicher Flächenwert der Viertelkreise über das Dreieck: {mean_area}")

```

```python
import numpy as np
from scipy.integrate import dblquad, quad

# Funktion zur Berechnung der Fläche der Viertelkreise an einem Punkt (x, y)
def calculate_quarter_circle_area(x, y):
    # Berechnung der Radien zu den Ecken bei x=0 und x=1
    corner_coords = [(0, 0), (1, 0)]  # Ecken bei x=0 und x=1 (untere linke und rechte Ecke)
    radii = []
    for corner in corner_coords:
        dx = x - corner[0]
        dy = y - corner[1]
        radius = np.sqrt(dx**2 + dy**2)
        radii.append(radius)

    # Funktion y_max(x) definieren
    def y_max(x):
        y1 = np.sqrt(max(radii[0]**2 - x**2, 0)) if x >= 0 and x <= radii[0] else 0
        y2 = np.sqrt(max(radii[1]**2 - (x - 1)**2, 0)) if x >= 1 - radii[1] and x <= 1 else 0
        return max(y1, y2, 0)

    # Numerische Integration durchführen
    area, _ = quad(y_max, 0, 1)
    return area

# Funktion für die Integration über das Dreieck
# Grenzen: x von 0 bis 1, und y von 0 bis 0.5 * x für das Dreieck
def integrand(x, y):
    return calculate_quarter_circle_area(x, y)

# Doppelintegration über das Dreieck
mean_area, _ = dblquad(integrand, 0, 1, lambda x: 0, lambda x: 0.5 * x)
# Gesamtfläche des Dreiecks ist 0.25, daher Mittelwert durch Division
mean_area /= 0.5

print(f"Exakter durchschnittlicher Flächenwert der Viertelkreise über das Dreieck: {mean_area:.10f}")
```

```python
import numpy as np
from scipy.integrate import quad

# Funktion zur Berechnung der Fläche der Viertelkreise an einem Punkt (x, y)
def calculate_quarter_circle_area(x, y):
    # Berechnung der Radien zu den Ecken bei x=0 und x=1
    corner_coords = [(0, 0), (1, 0)]  # Ecken bei x=0 und x=1 (untere linke und rechte Ecke)
    radii = []
    for corner in corner_coords:
        dx = x - corner[0]
        dy = y - corner[1]
        radius = np.sqrt(dx**2 + dy**2)
        radii.append(radius)

    # Funktion y_max(x) definieren
    def y_max(x):
        y1 = np.sqrt(max(radii[0]**2 - x**2, 0)) if x >= 0 and x <= radii[0] else 0
        y2 = np.sqrt(max(radii[1]**2 - (x - 1)**2, 0)) if x >= 1 - radii[1] and x <= 1 else 0
        return max(y1, y2, 0)

    # Numerische Integration durchführen
    area, _ = quad(y_max, 0, 1)
    return area

# Eckpunkte des Dreiecks
vertices = {
    "A": (0, 0),
    "B": (1, 0),
    "C": (0.5, 0.5)
}

# Berechnung der Fläche an den Eckpunkten
vertex_areas = {vertex: calculate_quarter_circle_area(x, y) for vertex, (x, y) in vertices.items()}

# Kanten des Dreiecks mit Berechnung der niedrigsten Werte in der Mitte jeder Kante
edges = [
    ("AB", vertices["A"], vertices["B"]),       # Kante AB von (0, 0) zu (1, 0)
    ("AC", vertices["A"], vertices["C"]),       # Kante AC von (0, 0) zu (0.5, 0.5)
    ("BC", vertices["B"], vertices["C"])        # Kante BC von (1, 0) zu (0.5, 0.5)
]

# Berechnung der Fläche in der Mitte jeder Kante
edge_mid_areas = {}
for edge_name, (x_start, y_start), (x_end, y_end) in edges:
    mid_x = (x_start + x_end) / 2
    mid_y = (y_start + y_end) / 2
    edge_mid_areas[edge_name] = calculate_quarter_circle_area(mid_x, mid_y)

# Berechnung der Fläche am Mittelpunkt des Dreiecks
center_x = (vertices["A"][0] + vertices["B"][0] + vertices["C"][0]) / 3
center_y = (vertices["A"][1] + vertices["B"][1] + vertices["C"][1]) / 3
center_area = calculate_quarter_circle_area(center_x, center_y)

# Gewichtete Mittelwertberechnung
weighted_sum = (
    1 * (vertex_areas["A"] + vertex_areas["B"] + vertex_areas["C"]) +
    2 * (edge_mid_areas["AB"] + edge_mid_areas["AC"] + edge_mid_areas["BC"]) +
    3 * center_area
)
total_weight = 1 * 3 + 2 * 3 + 3
weighted_mean_area = weighted_sum / total_weight

# Ausgabe der gewichteten Mittelwerte und des Gesamtergebnisses
print("Flächenwerte an den Eckpunkten:")
for vertex, area in vertex_areas.items():
    print(f"Punkt {vertex} -> Fläche: {area:.10f}")

print("\nNiedrigste Werte entlang der Kanten:")
for edge_name, mid_area in edge_mid_areas.items():
    print(f"Kante {edge_name} - Fläche in der Mitte: {mid_area:.10f}")

print(f"\nFläche am Mittelpunkt des Dreiecks: {center_area:.10f}")
print(f"\nDurchschnittliche Fläche der Viertelkreise über das Dreieck (gewichteter Mittelwert): {weighted_mean_area:.10f}")
```

```python
import numpy as np
from scipy.integrate import dblquad, quad

# Funktion zur Berechnung der Fläche der Viertelkreise an einem Punkt (x, y)
def calculate_quarter_circle_area(x, y):
    # Berechnung der Radien zu den Ecken bei (0, 0) und (1, 0)
    corner_coords = [(0, 0), (1, 0)]
    radii = []
    for corner in corner_coords:
        dx = x - corner[0]
        dy = y - corner[1]
        radius = np.sqrt(dx**2 + dy**2)
        radii.append(radius)

    # Bestimmung des maximalen y-Werts für den Punkt (x, y)
    def y_max(x):
        y1 = np.sqrt(max(radii[0]**2 - x**2, 0)) if x >= 0 and x <= radii[0] else 0
        y2 = np.sqrt(max(radii[1]**2 - (x - 1)**2, 0)) if x >= 1 - radii[1] and x <= 1 else 0
        return max(y1, y2, 0)

    # Integration entlang der y-Achse zwischen den beiden Viertelkreisen
    area, _ = quad(y_max, 0, 1)
    return area

# Definition des Integranden
def integrand(x, y):
    return calculate_quarter_circle_area(x, y)

# Integration über die beiden Bereiche des Dreiecks
mean_area1, _ = dblquad(integrand, 0, 0.5, lambda y: 0, lambda y: x)
mean_area2, _ = dblquad(integrand, 0.5, 1, lambda y: 0, lambda y: 1 - x)

# Gesamtmittelwert durch die Fläche des Dreiecks berechnen
mean_area = (mean_area1 + mean_area2) #/ 0.25
print(f"Exakter durchschnittlicher Flächenwert der Viertelkreise über das Dreieck: {mean_area:.10f}")

# Beispielhafte Berechnung der Fläche an ausgewählten Punkten
example_points = {
    "A (0, 0)": (0, 0),
    "B (1, 0)": (1, 0),
    "C (0.5, 0.5)": (0.5, 0.5),
    "Mitte der Basis (0.5, 0)": (0.5, 0),
    "Mitte von AB (0.25, 0)": (0.25, 0),
    "Mitte von AC (0.25, 0.25)": (0.25, 0.25),
    "Mitte von BC (0.75, 0.25)": (0.75, 0.25)
}

print("\nFlächenwerte an ausgewählten Punkten im Dreieck:")
for point_name, (x, y) in example_points.items():
    area = calculate_quarter_circle_area(x, y)
    print(f"Punkt {point_name} -> Fläche: {area:.10f}")
```

```python
import numpy as np
from scipy.integrate import dblquad

# Dreieckspunkte
A = (0, 0)
B = (1, 0)
C = (0, 1)

# Berechnung der Jacobi-Determinante
J = abs((B[0] - A[0]) * (C[1] - A[1]) - (C[0] - A[0]) * (B[1] - A[1]))

# Transformierte Funktion f(x, y) = x + y
def transformed_function(u, v):
    # Transformiere x und y basierend auf u und v
    x = A[0] + u * (B[0] - A[0]) + v * (C[0] - A[0])
    y = A[1] + u * (B[1] - A[1]) + v * (C[1] - A[1])
    # Gebe die Funktion und die Jacobi-Determinante zurück
    return (x + y) * J

# Grenzen für u und v
u_min, u_max = 0, 1
v_min = lambda u: 0
v_max = lambda u: 1 - u

# Integral berechnen
result, error = dblquad(transformed_function, u_min, u_max, v_min, v_max)
print("Das Integral ist:", result)
```

```python
import math
from fractions import Fraction

def intersection_area_exact(radius, distance):
    # calculate the area of intersection of two circles exactly
    theta = 2 * math.acos(distance / (2 * radius))
    area = 2 * (radius ** 2) * (theta / 2) - (distance / 2) * math.sqrt(4 * radius ** 2 - distance ** 2)
    
    # format area in terms of π using a fraction
    pi_coefficient = Fraction(area / math.pi).limit_denominator()
    return f"{pi_coefficient / 2} * π"

# Parameters
radius = math.sqrt(2) 

distance = 1

# Calculating the exact intersection area
print(intersection_area_exact(radius, distance))
```

```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt

# Koordinaten des Quadrats
square_x = [0, 1, 1, 0, 0]
square_y = [0, 0, 1, 1, 0]

# Koordinaten des Dreiecks
triangle_x = [0, 1, 0.5]
triangle_y = [0, 0, 0.5]

def plot_triangle_with_click():
    # Erstellen der Figur und der Achsen
    fig, ax = plt.subplots(figsize=(6,6))
    
    # Zeichnen des Quadrats
    ax.plot(square_x, square_y, 'k-')
    
    # Zeichnen des Dreiecks und Füllen
    ax.plot(triangle_x + [triangle_x[0]], triangle_y + [triangle_y[0]], 'b-')  # Dreieck schließen
    ax.fill(triangle_x, triangle_y, 'b', alpha=0.3)
    
    # Achsenlimits und -beschriftungen
    ax.set_xlim(-0.1, 1.1)
    ax.set_ylim(-0.1, 1.1)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Klicken Sie innerhalb des Dreiecks, um einen Punkt zu setzen')
    ax.set_aspect('equal', adjustable='box')
    ax.grid(True)
    
    # Liste zur Speicherung der gesetzten Punkte
    points_x = []
    points_y = []
    
    # Funktion zur Überprüfung, ob ein Punkt innerhalb des Dreiecks liegt
    def is_inside_triangle(x, y, triangle_coords):
        x1, y1 = triangle_coords[0]
        x2, y2 = triangle_coords[1]
        x3, y3 = triangle_coords[2]
        
        # Berechnung der baryzentrischen Koordinaten
        denominator = ((y2 - y3)*(x1 - x3) + (x3 - x2)*(y1 - y3))
        if denominator == 0:
            return False  # Dreieck ist degeneriert
        a = ((y2 - y3)*(x - x3) + (x3 - x2)*(y - y3)) / denominator
        b = ((y3 - y1)*(x - x3) + (x1 - x3)*(y - y3)) / denominator
        c = 1 - a - b
        
        return (0 <= a <= 1) and (0 <= b <= 1) and (0 <= c <= 1)
    
    # Event-Handler-Funktion für Mausklicks
    def onclick(event):
        if event.inaxes == ax:
            x_click = event.xdata
            y_click = event.ydata
            # Überprüfen, ob der Klick innerhalb des Dreiecks liegt
            triangle_coords = [(triangle_x[i], triangle_y[i]) for i in range(3)]
            if is_inside_triangle(x_click, y_click, triangle_coords):
                # Punkt speichern und zeichnen
                points_x.append(x_click)
                points_y.append(y_click)
                ax.plot(x_click, y_click, 'ro')  # 'ro' für roten Punkt
                # Optional: Koordinaten anzeigen
                ax.annotate(f"({x_click:.2f}, {y_click:.2f})", (x_click, y_click), textcoords="offset points", xytext=(5,5), ha='left')
                plt.draw()
            else:
                print("Der Klick liegt außerhalb des Dreiecks.")
    
    # Verbinden des Event-Handlers mit dem Figure-Canvas
    cid = fig.canvas.mpl_connect('button_press_event', onclick)
    
    plt.show()
    
plot_triangle_with_click()
```
